```python
import pandas as pd
import numpy as np
import altair as alt

alt.data_transformers.enable("default")

df = pd.read_csv("healthcare_cleaned.csv")

df['insurance_provider'] = df['insurance_provider'].str.strip().str.title()

df['insurance_provider'] = df['insurance_provider'].replace({
    "Unitedhealthcare": "United Healthcare",
    "Bluecross": "Blue Cross"})

df_sample = df.sample(n=1000, random_state=1)

df_sample['jitter'] = np.random.uniform(-0.25, 0.25, size=len(df_sample))

chart = alt.Chart(df_sample).mark_circle(
    size=40,
    opacity=0.4).encode(x=alt.X(
        'insurance_provider:N',
        title='Insurance Provider',
        axis=alt.Axis(labelAngle=0)),

    xOffset='jitter:Q',

    y=alt.Y('billing_amount:Q',title='Billing Amount ($)'),

    color=alt.Color('gender:N',title='Gender'),

    tooltip=[
        alt.Tooltip('insurance_provider:N', title='Provider'),
        alt.Tooltip('gender:N', title='Gender'),
        alt.Tooltip('age_group:N', title='Age Group'),
        alt.Tooltip('billing_amount:Q', title='Billing', format='$,.0f')]).properties(
            width=500,
            height=400,
            title='Billing Distribution by Insurance Provider and Gender (Sample of 1000)')


chart = chart.facet(
    column=alt.Column('gender:N', title='Gender')).resolve_scale(y='shared')

chart
#chart.save("billing_by_gender.png")
```

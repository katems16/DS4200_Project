```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("healthcare_cleaned.csv")

age_avg = (
    df.groupby(['age_group', 'insurance_provider'])['billing_amount']
    .mean()
    .reset_index()
    .rename(columns={'billing_amount': 'avg_billing'})
)

custom_colors = [
    "#4E79A7",  # blue
    "#F28E2B",  # orange
    "#59A14F",  # green
    "#E15759",  # red
    "#B07AA1",  # purple
    "#76B7B2"   # teal
]
```

```python
chart = (
    alt.Chart(age_avg, title="Average Billing by Age Group and Insurance Provider")
    .mark_bar(size=20, cornerRadiusTopLeft=3, cornerRadiusTopRight=3)
    .encode(
        x=alt.X(
            'age_group:O',
            title='Age Group',
            sort=['0–17','18–35','36–50','51–65','65+'],
            axis=alt.Axis(labelAngle=0)
        ),
        y=alt.Y(
            'avg_billing:Q',
            title='Avg Billing Amount ($)',
            axis=alt.Axis(grid=True)
        ),
        color=alt.Color(
            'insurance_provider:N',
            title='Insurance Provider',
            scale=alt.Scale(range=custom_colors),
            legend=alt.Legend(orient='top')
        ),
        xOffset='insurance_provider:N',
        tooltip=[
            alt.Tooltip('age_group:O', title='Age Group'),
            alt.Tooltip('insurance_provider:N', title='Provider'),
            alt.Tooltip('avg_billing:Q', title='Avg Billing', format='$,.0f')
        ]
    )
    .properties(width=700, height=400)
    .configure_title(
        fontSize=18,
        anchor='start',
        font='Arial'
    )
    .configure_axis(
        labelFontSize=12,
        titleFontSize=13
    )
    .configure_legend(
        titleFontSize=12,
        labelFontSize=11
    )
)

chart.show()
chart.save("Billing By Age Group and Insurance.png")
```

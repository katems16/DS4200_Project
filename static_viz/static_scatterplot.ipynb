```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("healthcare_cleaned.csv")

age_avg = (
    df.groupby(['age_group', 'insurance_provider', "gender"])['billing_amount']
    .mean()
    .reset_index()
    .rename(columns={'billing_amount': 'avg_billing'})
)

custom_colors = [
    "#FF6BD2",      # pink
    "#BA6BFF",      # purple
    "#6B6BFF",      # blue
    "#70FF6B",      # green
    "#FFCE6B",      # orange
    "#F5FF6B"       # yellow
]
```

```python
base = (
    alt.Chart(age_avg, title="Comparing Average Billing by Gender")
    .mark_point(size=60, filled=True)
    .encode(
        x=alt.X(
            'insurance_provider:O',
            title='Insurance Provider',
            sort=['aetna', 'blue cross', 'cigna', 'medicare', 'unitedhealthcare'],
            axis=alt.Axis(labelAngle=0)
        ),
        y=alt.Y(
            'avg_billing:Q',
            title='Avg Billing Amount ($)',
            axis=alt.Axis(grid=True)
        ),
        color=alt.Color(
            'age_group:N',
            title='Age Group',
            scale=alt.Scale(range=custom_colors),
            legend=alt.Legend(orient='top')
        ),
        tooltip=[
            alt.Tooltip('age_group:N', title='Age Group'),
            alt.Tooltip('insurance_provider:O', title='Provider'),
            alt.Tooltip('gender:N', title='Gender'),
            alt.Tooltip('avg_billing:Q', title='Avg Billing', format='$,.0f'),
        ]
    )
)

chart = (
    base
    .properties(width=500, height=400)
    .facet(
        facet=alt.Facet('gender:N', title='Gender'),
        columns=2
    )
    .resolve_scale(y='shared')
    .properties(
        title=alt.TitleParams(
            text='Comparing Average Billing by Gender',
            fontSize=18,
            anchor='start',
            font='Arial'
        )
    )
    .configure_axis(
        labelFontSize=12,
        titleFontSize=13
    )
    .configure_legend(
        titleFontSize=12,
        labelFontSize=11
    )
    .configure_header(
        titleFontSize=14,
        labelFontSize=13,
        labelFont='Arial'
    )
)

chart.show()
chart.save("Comparing Average Billing by Gender.png")
```

### Loaded Dataframe (not fully cleaned/processed):

```python
import pandas as pd
import numpy as np
df = pd.read_csv('healthcare_dataset.csv')

df['Billing Amount']= pd.to_numeric(df['Billing Amount'],errors='coerce')
df['Billing Amount'] = df['Billing Amount'].fillna(0.0)
df['Insurance Provider'] = df['Insurance Provider'].replace('UnitedHealthcare','United Healthcare')

df.head()
```

### Checkpoint Visualizations

#### 1) Bar chart: insurance providers and their average billing costs for arthritis

```python
import plotly.express as px
import statistics

avg_billing = df.groupby('Insurance Provider')['Billing Amount'].mean().reset_index()

arth_df = df[df['Medical Condition'] == 'Arthritis']
avg_billing_arth = (arth_df.groupby('Insurance Provider')['Billing Amount'].mean().reset_index())


fig1 = px.bar(avg_billing_arth, 
              x='Insurance Provider', 
              y='Billing Amount', 
              color='Insurance Provider',
              title='Insurance Providers and their Average Billing Costs for Arthritis',
             width=1000,
            height=600)

fig1.update_yaxes(range=[25000, 26000])
fig1.update_yaxes(title='Average Billing Cost ($)')

fig1.show()
```

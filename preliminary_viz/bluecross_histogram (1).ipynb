```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()

df = pd.read_csv('healthcare_dataset.csv')
```

```python
df_blue_cross = df[df['Insurance Provider'] == 'Blue Cross']

billing_mean = df_blue_cross['Billing Amount'].mean()
print(f"Mean Billing Amount for Blue Cross: {billing_mean}")

plt.figure(figsize=(10, 6))
sns.histplot(data=df_blue_cross, x='Billing Amount', kde=True)
plt.title('Distribution of Billing Amount for Blue Cross')
plt.show()
```

```python
provider_means = df.groupby("Medical Condition")["Billing Amount"].mean()
overall_mean = df["Billing Amount"].mean()
diff = (provider_means - overall_mean).sort_values()

plt.figure(figsize=(10, 6))

colors = ["red" if v < 0 else "green" for v in diff]

plt.barh(diff.index, diff.values, color=colors)
plt.axvline(0)

plt.title("Deviation from Average Billing by Medical Condition")
plt.xlabel("Difference from Average Billing Amount")
plt.ylabel("Medical Condition")
plt.tight_layout()
plt.show()
```

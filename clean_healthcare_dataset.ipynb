```python
import pandas as pd
import numpy as np

df = pd.read_csv("healthcare_dataset.csv")
```

```python
# Normalize the columns
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df['name'] = df['name'].str.lower().str.replace(r'\b(mr\.|mrs\.|ms\.|dr\.)\s*', '', regex=True).str.strip()
df[['first_name', 'last_name']] = df['name'].str.split(' ', n=1, expand=True)

df['medical_condition'] = df['medical_condition'].str.lower().str.strip()
df['insurance_provider'] = df['insurance_provider'].str.lower().str.strip()
df['hospital'] = df['hospital'].str.lower().str.strip()
df['admission_type'] = df['admission_type'].str.lower().str.strip()

# Parse dates and compute length of stay
df['date_of_admission'] = pd.to_datetime(df['date_of_admission'])
df['discharge_date'] = pd.to_datetime(df['discharge_date'])
df['length_of_stay'] = (df['discharge_date'] - df['date_of_admission']).dt.days
df['year'] = df['date_of_admission'].dt.year
df['billing_amount'] = pd.to_numeric(df['billing_amount'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# Age groups
bins = [0, 17, 35, 50, 65, 120]
labels = ['0–17', '18–35', '36–50', '51–65', '65+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Daily billing rate
df['daily_billing'] = df['billing_amount'] / df['length_of_stay']

df.to_csv("healthcare_cleaned.csv", index=False)
print(df.shape)
print(df.head())
```
